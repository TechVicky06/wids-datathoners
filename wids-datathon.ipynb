Align all datasets by participant_id to ensure there are no mismatches (e.g., ensure all participants in training_solutions exist in the metadata files).

```python
#heck for missing values in each dataset
print("\nMissing Values:")
print("Quantitative Metadata:", train_quantitative.isnull().sum())
print("Categorical Metadata:", train_categorical.isnull().sum())
print("Training Solutions:", train_solutions.isnull().sum())

# Align datasets by participant_id
# Check if all participant IDs in training_solutions are present in the other datasets
quant_ids = set(train_quantitative['participant_id'])
cat_ids = set(train_categorical['participant_id'])
sol_ids = set(train_solutions['participant_id'])

print("\nParticipant ID Alignment:")
print("IDs in Quantitative Metadata but not in Solutions:", quant_ids - sol_ids)
print("IDs in Categorical Metadata but not in Solutions:", cat_ids - sol_ids)
print("IDs in Solutions but not in Quantitative Metadata:", sol_ids - quant_ids)
print("IDs in Solutions but not in Categorical Metadata:", sol_ids - cat_ids)

# Align all datasets by intersecting participant IDs
common_ids = quant_ids & cat_ids & sol_ids
train_quantitative = train_quantitative[train_quantitative['participant_id'].isin(common_ids)]
train_categorical = train_categorical[train_categorical['participant_id'].isin(common_ids)]
train_solutions = train_solutions[train_solutions['participant_id'].isin(common_ids)]

print("\nAfter Alignment:")
print(f"Quantitative Metadata: {train_quantitative.shape}")
print(f"Categorical Metadata: {train_categorical.shape}")
print(f"Training Solutions: {train_solutions.shape}")
```

```python
# Merge the datasets for analysis
merged_data = train_quantitative.merge(train_solutions, on='participant_id', how='inner')

# Filter for numeric columns
numeric_data = merged_data.select_dtypes(include=[np.number])

# Check correlation
correlations = numeric_data.corr()

# Ensure the column 'MRI_Track_Age_at_Scan' exists in the numeric data
if 'MRI_Track_Age_at_Scan' in correlations.index:
    mri_correlation_with_targets = correlations.loc['MRI_Track_Age_at_Scan', ['ADHD_Outcome', 'Sex_F']]
    print("Correlation of MRI_Track_Age_at_Scan with target variables:\n", mri_correlation_with_targets)
else:
    print("MRI_Track_Age_at_Scan column not found in numeric data.")

# Visualization
plt.figure(figsize=(12, 6))

# Boxplot for ADHD_Outcome
plt.subplot(1, 2, 1)
sns.boxplot(data=merged_data, x='ADHD_Outcome', y='MRI_Track_Age_at_Scan')
plt.title('MRI_Track_Age_at_Scan vs. ADHD_Outcome')

# Boxplot for Sex_F
plt.subplot(1, 2, 2)
sns.boxplot(data=merged_data, x='Sex_F', y='MRI_Track_Age_at_Scan')
plt.title('MRI_Track_Age_at_Scan vs. Sex_F')

plt.tight_layout()
plt.show()

# Histogram
plt.figure(figsize=(6, 4))
sns.histplot(data=merged_data, x='MRI_Track_Age_at_Scan', kde=True, bins=20)
plt.title('Distribution of MRI_Track_Age_at_Scan')
plt.xlabel('MRI_Track_Age_at_Scan')
plt.ylabel('Frequency')
plt.show()
```

Low Correlation Values: Both correlations are close to zero, indicating that MRI_Track_Age_at_Scan likely does not contribute significantly to predicting either ADHD or sex based on a linear relationship.
